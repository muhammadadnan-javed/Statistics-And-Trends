```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("dummy_data.csv")
data.head()
```

```python
data.describe()
```

```python
data.corr()
```

```python
sns.set_style("whitegrid")

# Relational Graph: Scatter plot of age vs. time_spent
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='age', y='time_spent', data=data, hue='platform', style='gender', s=100)
scatter_plot.set_title('Age vs. Time Spent on Platforms', fontsize=16)
scatter_plot.set_xlabel('Age', fontsize=14)
scatter_plot.set_ylabel('Time Spent (hours)', fontsize=14)
plt.legend(title='Platform and Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# Categorical Graph: Bar chart of platform usage across demographics
plt.figure(figsize=(10, 6))
bar_chart = sns.countplot(x='platform', hue='demographics', data=data, palette='Set2')
bar_chart.set_title('Platform Usage Across Demographics', fontsize=16)
bar_chart.set_xlabel('Platform', fontsize=14)
bar_chart.set_ylabel('Count', fontsize=14)
plt.legend(title='Demographics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
# Statistical Graph: Box plot of income across professions
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='income', y='profession', data=data, palette='coolwarm')
box_plot.set_title('Income Distribution Across Professions', fontsize=16)
box_plot.set_xlabel('Income', fontsize=14)
box_plot.set_ylabel('Profession', fontsize=14)
plt.tight_layout()

plt.show()
```
